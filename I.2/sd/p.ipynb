```python
#1
t=var('t')
Tm=var('Tm')
k=var('k')
T0=var('T0')
T=function('T')(t)
eq=diff(T,t)== -k * (T(t) - Tm)
sol = desolve(eq,T,ivar=t,ics=[0,T0])
sol


sol2 = sol.substitute(k==0.5)
sol2 = sol2.substitute(Tm==45)
sol2 = sol2.substitute(T0==25)
sol2

plot(sol2,T,-10,10)
```
